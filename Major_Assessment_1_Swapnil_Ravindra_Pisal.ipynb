```python
# Q3. Function and Logic Create a user-defined function that:
# Accepts a number as input
# Returns whether the number is a Palindrome number or not

def pallindrome(num):
  if num==num[::-1]:
    print("It is a pallindrome number")
  else:
    print("It is a not pallindrome number")

pallindrome("123")
pallindrome("121")
pallindrome("Swapnil")
pallindrome("nitin")

```

# Section B - NumPy  

```python
# Q4. NumPy Array Operations
import numpy as np
```

```python
 #  Create a NumPy array of integers from 1 to 20
num_arr = np.arange(1,21)
print("The NumPy array of integers from 1 to 20 :\n",num_arr)
```

```python
# Reshape it into a 4 × 5 matrix

num_arr_reshape=num_arr.reshape(4,5)

print("Reshape it into a 4 × 5 matrix are : \n",num_arr_reshape)
```

```python
# Find:
# Mean of each column
# Maximum value of each row


print("Mean of each column : ",np.mean(num_arr_reshape,axis=0))
print("Maximum value of each row : ",np.max(num_arr_reshape,axis=1))

```

```python
# Q5. Conditional Operations Using NumPy:
# Replace all values greater than 10 with -1
# Display the modified array

modified_arr=np.where(num_arr_reshape>10,-1,num_arr_reshape)
print("The Modified Array are : \n",modified_arr)

```

```python
# Handle missing values appropriately

# To chesk the missing values
print(df.isnull().sum())
```

```python
# fill missing values of numeric column

df['retail_price'].fillna(df['retail_price'].mean(),inplace=True)

df['discounted_price'].fillna(df['discounted_price'].mean(),inplace=True)
```

```python
# To check after filling values of numerical column

df.isnull().sum()
```

```python
# To fill all object column using Mode method

df['uniq_id'].fillna(df['uniq_id'].mode()[0],inplace=True)

df['crawl_timestamp'].fillna(df['crawl_timestamp'].mode()[0],inplace=True)

df['product_url'].fillna(df['product_url'].mode()[0],inplace=True)

df['product_name'].fillna(df['product_name'].mode()[0],inplace=True)

df['product_category_tree'].fillna(df['product_category_tree'].mode()[0],inplace=True)

df['pid'].fillna(df['pid'].mode()[0],inplace=True)

df['image'].fillna(df['image'].mode()[0],inplace=True)

df['is_FK_Advantage_product'].fillna(df['is_FK_Advantage_product'].mode()[0],inplace=True)

df['description'].fillna(df['description'].mode()[0],inplace=True)

df['product_rating'].fillna(df['product_rating'].mode()[0],inplace=True)

df['overall_rating'].fillna(df['overall_rating'].mode()[0],inplace=True)

df['brand'].fillna(df['brand'].mode()[0],inplace=True)

df['product_specifications'].fillna(df['product_specifications'].mode()[0],inplace=True)
```
